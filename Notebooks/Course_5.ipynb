# The electromotive "force"

During course 4, we digged deeper into the first circuit I presented you: the **light bulb/battery circuit**. Thanks to Ohm's law, we understood that an electric field $E$ was present into the light bulb filament (modeled as a **resistor** of resistance $R$) due to a difference of potential $\Delta V$ imposed by the battery. As a **consequence**, electric charge flows throught the resistor and the electric field and the current density $J$ are linked by the following equation:

$$\rho=\frac{E}{J}$$

We also know (from INGE1243) that the electric field is a **vector field** whose direction is given as **positive** when going from the posive charges towards negative charges. You also know that, due to charge conservation, the current $I$ is the same no matter where it is measured in the circuit. Here is a drawing summing up these observations:

<img title="a title" alt="Alt text" src="../Figures/Cours 5/battery.png" width="400">

From the electric point of view, this picture is correct. You should now be familiar with the current flowing throught the resistor. But the current flowing throught the battery (the same as throught the resistor) could bother you. You might wonder **how** a current could be maintained and flowing in the opposite direction of the electric field $E$. This is a legitimate concern since it is exactly as if a ball was climbing up a hill even though gravity is pushing it from higher to lower potential energy. The influence that makes current flow from lower to higher potential is called the **electromotive force** (abbraviated **emf**). Every complete circuit in which you can measure a steady current **must** include a device acting a a source of **emf**. Such device turns any source of energy (chemical, thermal, mechanical, etc.) into electric **potential energy** ($\Delta V$). The role of this emf source is to **impose** a difference of potential between the terminals (input/output) of an electrical device. For example, in the above circuit, the battery **imposes** a **constant** difference of potential between its terminals, and, if the source is **ideal**, it does so, **no matter the current through it**.

Suppose I plug a 2 $\Omega$ resistor onto a 12V battery, a current of 6A is flowing through the circuit. If I now plug a 4 $\Omega$ resistor to the same battery, **ideally**, I get a current of 3A. Eventhough the current **through the battery** is now halved, the difference of potential between its two terminal remains 12V (a little bit magical isn't it ?). It is indeed not exactly how these object work, but modern days battery are getting closer and closer to this **ideal** model. 

The ideal situation is described by the following figure:

<img title="a title" alt="Alt text" src="../Figures/Cours 5/graph.png" width="800">

The figure presents a simple battery/resistor circuit (left part of the figure). An orange dashed line represents a **virtual** path on which the evolution of the electric potential varies. Starting from the white dot (corresponding to the **negative** terminal of the battery), the path goes **through** the battery, then through the **wire** connecting the **positive** terminal to the first connector of the resistor, then through the resistor and eventually back to the negative terminal ($s$ - the abscissa of the right graph) represents the main parameter (units of length) of the closed path. 

There are 4 distinct parts on this graph, the first on is the evolution of the electric potential through the battery. The potential **rises** although the electric field is directed from the "+" to the "-" terminal. A charge $q$ within the battery experiences an electric force $\vec{F}_{e}=q\vec{E}$. **But** the charge experiences another **non-electrostatic** force $\vec{F}_{n}$. If this force was not present, the battery would simply discharge. The **nature** of $\vec{F}_{n}$ would be properly described in an **electrochemistry** course (as it is not our main purpose, we will not dig too deep into this question). In a battery, this **work** is associated with **diffusion processes** and **varying electrolyte concentrations** resulting from chemical reactions (more about this at the end of this notebook). In the battery, we have:

<img title="a title" alt="Alt text" src="../Figures/Cours 5/forces.png" width="200">

As the charge goes from "-" to "+", $\vec{F}_{n}$ realizes a **work** $W_{n}$ given by:

$$W_{n}=q\mathcal{E}$$

where $\mathcal{E}$ is designated as the **emf**. As you see, this **emf** (electromotive "force") is **not** a force as it has the dimension of energy per electric charge. The SI unit of the emf is joules per coulomb $[J/C]$.

On the other hand, the work associated to the electric field is given by:

$$W_{e}=qV$$

where $V$ is the difference of potential between the two terminals of the battery. 

<div class="alert alert-block alert-info">
Note that $V$ is potential **difference**. To measure a difference, it is fundamental to define a **reference potential**. It is done with the help of the **ground symbol** (single asterisk on the figure presenting the circuit). This symbol signifies that every value of potential expressed in the circuit is taken as a difference with respect to the "ground potential" - equal to "0" by convention. 

The difference of potential $\Delta V$ is simply noted $V$ ($=\Delta V=V-0$).
</div>

The important thing to notice is that:

**These two works are exactly equals**

Indeed, if one was greater to the other, a net work would be realized on each charge going through the battery. If this was the case, thanks to the **work-energy theorem**, these charges would gain **kinetic energy** ($\Delta K.E.=work$). But as we saw in the previous course, a change in the potential in the circuit is associated to a current **defined** by a **constant** velocity (e.g. the drift velocity in the resistor). Changing **only** the potential is just as lifting a bag of ciment from the first floor to the second (as done by Tom or Jerry in course 7 of INGE1243).

Since both works are equals, we have:

$$W_{n}=W_{e}\leftrightarrow V=\mathcal{E}$$

for the **ideal** source of emf.

As the resistance is connected to the battery, we know that (thanks to Ohm's law):

$$V=RI=\mathcal{E}$$

This equation is not value for any **real world** source of emf. Indeed, these sources always present **resistance** to the current flow. This resistance is called an **internal resistance** noted with the letter $r$. In most cases, this resistance can be considered as independent of the current $I$ and modeled as a constant resistor (expressed in ohms). 

The potential $V$ present between the two battery terminals is **always less** than the emf of the battery because a current $I$ flowing through the internal resistor causes a voltage drop equal to $Ir$.

The **ideal** voltage becomes:

$$V=\mathcal{E}-Ir$$

the actual current $I$ flowing through the resistor is then given by:

$$V=IR=\mathcal{E}-Ir\leftrightarrow I=\frac{\mathcal{E}}{R+r}$$

To go further in our comprehension, let us consider 2 peculiar situations: the open circuit and the short circuit.

<img title="a title" alt="Alt text" src="../Figures/Cours 5/openshort.png" width="500">

In the open circuit configuration, no current flows through the circuit and thus, $$V=\mathcal{E}-0$$. In the short circuit configuration, the value $R$ equals to zero. But there is still an internal resistance $r$. Which limits the maximal current that the battery can possibly deliver to:

$$I_{max}=\frac{\mathcal{E}}{r}$$

The power is, in this short cutted configuration, **entirely** dissipated inside the battery itself. This power is expressed as:

$$P=VI=\mathcal{E}\times\frac{\mathcal{E}}{r}$$

Let us consider an actual car battery. As you probably know, they are about 12V (emf - $\mathcal{E}$) and (lesser known) with an internal resistance (r) of about $\frac{1}{50}$ ohms. The dissipated power is then 7200 W. As a comparison, a kitchen oven delivers an average power of 3000 W to maintain a constant 200°C temperature. This value of 7200 W is huge.

# The battery 'back of the envelope' working principle

This topic will be extensivelly covered during the ex cathedra lecture. 

Here is a figure illustrating the realization of a battery cell based on copper (Cu) and magnesium (Mg) (in your reference book, you could find the working principle of a lead-acid cell). 

<img title="a title" alt="Alt text" src="../Figures/Cours 5/battery_chemistry.png" width="700">

To start with the realization of this battery, one needs to realize that some chemical reactions **need** energy and other **deliver** energy. Could we benefit from this natural tendency? Yes... It is the principle of the battery. 

On the one hand, we find solid magnesium ($Mg$) that dissolves by releasing to electrons:

$$Mg_{s}\rightarrow Mg^{2+}_{aq}+2e^{-}$$

The potential energy acquired by the electrons in this chemical reaction is best described in the framework of quantum chemistry but the important thing to note is that it is **uniquely** determined by the chemicals. For example, in the standard conditions (at a temperature of 25°C and a $Mg^{2+}$ solution of concentration of 1 molar), this reaction (redox equilibrium) leads to **-2.372 V**.

On the other hand, we find a solution that contains copper ions ($Cu^{2+}_{aq}$) that, when capturing two electrons, becomes solid copper:

$$Cu_{aq}^{2+}+2e^{-}\rightarrow Cu_{s}$$

This reaction (redox equilibrium) leads to **+0.345 V**.

If, in a battery, on can maintain these two reactions, an **emf** of $\mathcal{E}=0.345-(-2.372)=2.717 V$.
