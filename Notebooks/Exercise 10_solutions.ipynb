### Question 1

A frequency $f$ is always given as the inverse of a period $T$:

$$f=\frac{1}{T}$$

The period of the relaxation oscillator is:

$$T_{osc}=2ln(2)RC$$

Thus the frequency of the oscillator is:

$$f_{osc}=\frac{1}{2ln(2)RC}$$

which will be approximated (read the note) to:

$$f_{osc}=\frac{1}{RC}$$

### Question 2

The filter presented in (a) is a **low pass filter**, signifying that it **transmits low frequencies** - frequencies lower than the cut-off frequency (noted $f_{cut}$) are **"OK"**), and **attenuates high frequencies** (frequencies above $f_{cut}$ are **"KO"**).

The filter presented in (b) is a **high pass filter**, signifying that it **transmits high frequencies** (higher than $f_{cut}$), and **attenuates low frequencies** (lower than $f_{cut}$).

The cut-off frequency of the oscillator is:

$$f_{cut}=\frac{1}{R^{*}C^{*}}$$

### Question 3

The strategy is first to compute $f_{osc}$:

$$f_{osc}=\frac{1}{RC}=\frac{1}{10\times10^{3}\times1\times10^{-9}}[s^{-1}]=0.1\times10^{6}\;[Hz]$$

then, to compute the cut-off frequency of the filter $f_{cut}$:

$$f_{cut}=\frac{1}{R^{*}C^{*}}=\frac{1}{10\times10^{3}\times1\times10^{-12}}[s^{-1}]=100\times10^{6}\;[Hz]$$

The conclusion goes as follow: the frequency of the oscillator is way smaller than the cut-off frequency of the filter ($f_{osc}<f_{cut}$). Since it is a **low pass filter**, the signal coming from the oscillator is in the "OK" region of the filter and so, it is transmitted.
